```python
for index, row in df.iterrows():

    raw_prompt = row["problem_statement"]

    scaffold_prompt = f"""
Write a Python solution suitable for a first-year programming student.

Problem:
{row['problem_statement']}

Instructions:

1. Use meaningful variable names.
2. Use simple loops.
3. Avoid slicing.
4. Avoid lambda.
5. Avoid list comprehensions.
6. Add comments.
7. Explain each major step.
8. Return the solution.
"""

    with open(f"prompts/raw/{row['problem_id']}.txt","w") as f:
        f.write(raw_prompt)

    with open(f"prompts/scaffolded/{row['problem_id']}.txt","w") as f:
        f.write(scaffold_prompt)

print("Prompt files created successfully.")
```
