````python
# ============================================================
# Cell 8 : Generate Raw Prompt Files
# ============================================================

import pandas as pd

dataset_file = DATASET_DIR / "problems.csv"
df = pd.read_csv(dataset_file)

RAW_PROMPTS_DIR.mkdir(parents=True, exist_ok=True)

for _, row in df.iterrows():

    prompt = f"""# Problem ID
{row['problem_id']}

# Title
{row['title']}

# Category
{row['category']}

# Difficulty
{row['difficulty']}

# Programming Task

{row['problem_statement']}

## Input
{row['input_description']}

## Output
{row['output_description']}

## Sample Input

{row['sample_input']}

## Sample Output

{row['sample_output']}

## Constraints

{row['constraints']}

### Instruction

Write a correct Python program that solves the above problem.

Return ONLY Python code.

Do not include any explanation.

Do not include markdown.

Do not include comments unless necessary.
"""

    filename = RAW_PROMPTS_DIR / f"{row['problem_id']}.md"

    with open(filename, "w", encoding="utf-8") as f:
        f.write(prompt)

print(f"Generated {len(df)} raw prompt files.")

print("\nPrompt Folder:")
print(RAW_PROMPTS_DIR)
````

````python
# ============================================================
# Cell 9 : Verify Prompt Files
# ============================================================

prompt_files = sorted(RAW_PROMPTS_DIR.glob("*.md"))

print("=" * 60)
print("Prompt File Verification")
print("=" * 60)

print(f"Number of Prompt Files : {len(prompt_files)}")

print("\nFiles:")

for file in prompt_files:
    print(file.name)

print("\nPreview of First Prompt\n")

with open(prompt_files[0], "r", encoding="utf-8") as f:
    print(f.read())
````
